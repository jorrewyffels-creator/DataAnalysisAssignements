## Introduction and Naive Interpretation
The initial linear regression analysis revealed two unexpected relationships that shaped our subsequent hypothesis testing: a strong negative correlation between Physically Active Hours and Grades, and a strong positive correlation between Stress Level and Grades. This suggested two possibilities: that exercise directly harms academic performance, and that a high stress level improves cognitive function.

This led to our two critical null hypotheses:

Activity Hypothesis: "Physical Activity has no significant impact on Grades when controlling for Study Hours."

Stress Hypothesis: "Stress has no significant impact on Grades when controlling for Study Hours."

Our goal in the following figure is to test whether the effects seen in our initial plots are real, or whether they are entirely explained by the mediating factor of Study Hours.

## Conclusion and Insightful Analysis
The controlled analysis, presented above, definitively resolves both issues, proving that Study Hours are the single, central driver of academic success.For the Physical Activity analysis (top panel), we clearly see that the slopes flattened out. For all study groups, the regression line is near zero or extremely shallow. Even for the high study group, the slope is dramatically reduced compared to the naive figure. The negative correlation vanishes because we explained it away: time lost to the gym is the real problem, not the gym itself.The Stress analysis (bottom panels) completely kills the "good stress" myth. In all study groups where a comparison is possible (Low Study, Average Study), the T-tests/ANOVA showed no significant difference in grades between relaxed and highly stressed students ($p > 0.05$). This confirms that the apparent grade advantage of stress was spurious: it wasn't stress driving the grades, it was the extra study time causing the stress.This proves how important it is to do insightful analysis instead of just naively interpreting your results.
