### Figure 2

Use `data-fig2.npy` for this figure.

```python
b   = np.load( 'data-fig2.npy' )

print(b.shape)
plt.figure()
plt.plot(b[:,0], c = "red", label='Red')
plt.plot(b[:,1], c = "Green", label='Green')
plt.plot(b[:,2], c = "Blue", label='Blue')
plt.legend()
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c   = np.load( 'data-fig3.npy' )

plt.figure()
plt.scatter( c[:,0], c[:,1], color='blue', alpha=0.3, s=200)
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d   = np.load( 'data-fig4.npy' )
xlabels = ['A','B','C','D','E']
x = [0, 1, 2, 3, 4]

plt.figure()
plt.bar( range(len(d)), d, color='.7')
plt.xlabel("Category")
plt.ylabel("Value")
plt.xticks( ticks=x, labels=xlabels , size=9 )
plt.show()
```
