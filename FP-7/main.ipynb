**Result 1: Study Hours Dominates Stress Prediction**

The Decision Tree model reveals that **study hours (72.4%)** is by far the most important predictor of stress, with **sleep hours (27.6%)** as secondary.  Social time, exercise, and gender have zero importance.

**Key insight:** Once you know a student's study hours and sleep hours, knowing their social time or exercise habits doesn't help predict stress any better.  This suggests that **academic pressure is the overwhelming source of student stress**, and interventions should prioritize study-life balance and adequate rest.

**Result 2: Decision Tree Achieves Perfect Classification Across All Folds**

The Decision Tree achieved **100% accuracy across all 5 cross-validation folds** (0 errors out of 2000 predictions) while KNN achieved approximately **91.8% accuracy** (~164 errors).

**Where KNN fails:** KNN's errors are concentrated at class boundaries.  It confuses Moderate stress with Low (~44 errors) and Moderate with High (~55 errors). This happens because KNN averages the nearest neighbors, which doesn't work well when classes have sharp boundaries.

**Why Decision Tree wins:** This dataset has clear, deterministic rules.  The tree learned exact thresholds that perfectly separate stress categories. These results are **robust across all 5 random train/test splits**, confirming this is not overfitting to a lucky split.

**Key insight:** The Decision Tree's perfect accuracy across all folds reveals that stress levels in this dataset follow deterministic rules based on study and sleep hours.
